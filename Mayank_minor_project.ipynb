```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. डेटा जनरेशन
np.random.seed(42)
n_samples = 10000
area = np.random.normal(2000, 500, n_samples)
bedrooms = np.random.randint(1, 6, n_samples)
age = np.random.uniform(0, 50, n_samples)
price = 150 * area + 10000 * bedrooms - 500 * age + np.random.normal(0, 20000, n_samples)

df = pd.DataFrame({'area': area, 'bedrooms': bedrooms, 'age': age, 'price': price})

# 2. 5% डेटा हटाना (Missing Values)
for col in df.columns[:-1]:
    df.loc[df.sample(frac=0.05).index, col] = np.nan

    # 3. डेटा क्लीनिंग (Mean Imputation)
    df.fillna(df.mean(), inplace=True)

    # 4. स्केलिंग (Standardization)
    df_scaled = (df - df.mean()) / df.std()
    print("डेटा तैयार है और स्केल कर दिया गया है।")
```

```python
class SimpleLR:
    def __init__(self, lr=0.1, iters=100):
        self.lr, self.it = lr, iters
        self.m = self.c = 0

    def fit(self, X, y):
        n = len(X)
        for _ in range(self.it):
            y_pred = self.m * X + self.c
            self.m -= self.lr * (-2/n) * sum(X * (y - y_pred))
            self.c -= self.lr * (-2/n) * sum(y - y_pred)

model1 = SimpleLR()
model1.fit(df_scaled['area'], df_scaled['price'])
print(f"Part 1 पूरा हुआ। Slope (m): {model1.m:.2f}")
```

```python
class MultipleLR:
    def fit(self, X, y):
        # इंटरसेप्ट के लिए 1 की कॉलम जोड़ना (Matrix math के लिए ज़रूरी)
        X_b = np.c_[np.ones((len(X), 1)), X]
        # Normal Equation फॉर्मूला: theta = (XT * X)^-1 * XT * y
        self.theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

    def predict(self, X):
        X_b = np.c_[np.ones((len(X), 1)), X]
        return X_b.dot(self.theta)

# मल्टीपल फीचर्स चुनना
X_mult = df_scaled[['area', 'bedrooms', 'age']]
y_mult = df_scaled['price']

model2 = MultipleLR()
model2.fit(X_mult, y_mult)
print("Part 2 (Multiple Features) सफलतापूर्वक पूरा हुआ!")
```

```python
# Part 3: Polynomial (areaarea का स्क्वायर जोड़ना)
df_poly = df_scaled.copy()
df_poly['area2'] = df_scaled['area']**2
X_poly = df_poly[['area', 'area2', 'bedrooms', 'age']]
model3 = MultipleLR()
model3.fit(X_poly, y_mult)

# मॉडल की तुलना के लिए RMSE फंक्शन
def get_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

# सबकी तुलना करना
rmse1 = get_rmse(y_mult, model1.m * df_scaled['area'] + model1.c)
rmse2 = get_rmse(y_mult, model2.predict(X_mult))
rmse3 = get_rmse(y_mult, model3.predict(X_poly))

print(f"--- फाइनल प्रोजेक्ट रिपोर्ट ---")
print(f"Simple Model Error (RMSE): {rmse1:.4f}")
print(f"Multiple Model Error (RMSE): {rmse2:.4f}")
print(f"Polynomial Model Error (RMSE): {rmse3:.4f}")
```
